#### Handling all the `-` together by replacing them with `0`

```python
odi_bat_df = odi_bat_df.replace('-', '0')
odi_bat_df
```

#### checking the data-types

```python
odi_bat_df.info()
```

#### `HS` column will be handled separately as it has a `*` in it. For now, first let us change the data types of columns
#### Also, `Span` column will also be handled separately

```python
odi_bat_df['Inns'] = odi_bat_df['Inns'].astype(int)
odi_bat_df['NO'] = odi_bat_df['NO'].astype(int)
odi_bat_df['Runs'] = odi_bat_df['Runs'].astype(int)
odi_bat_df['Ave'] = odi_bat_df['Ave'].astype(float)
odi_bat_df['BF'] = odi_bat_df['BF'].astype(int)
odi_bat_df['SR'] = odi_bat_df['SR'].astype(float)
odi_bat_df['100'] = odi_bat_df['100'].astype(float)
odi_bat_df['50'] = odi_bat_df['50'].astype(float)
odi_bat_df['0'] = odi_bat_df['0'].astype(float)

odi_bat_df.info()
```

#### Renaming column for better understanding

```python
odi_bat_df.rename(columns={
    'Span': 'Span_ODI_bat',
    'Mat': 'Mat_ODI_bat',
    'Inns': 'Inns_ODI_bat',
    'NO': 'NO_ODI',
    'Runs': 'Runs_ODI_bat',
    'HS': 'HS_ODI',
    'Ave': 'Ave_ODI_bat',
    'BF': 'BF_ODI',
    'SR': 'SR_ODI_bat',
    '100': '100_ODI',
    '50': '50_ODI',
    '0': '0_ODI'
}, inplace=True)

odi_bat_df
```

#### Let us now first handle `Span` column
What we can do is to seprate both and make two new columns of `Start year` and `End year`

```python
start = []
end = []
for span in odi_bat_df['Span_ODI_bat']:
    s_year, e_year = span.split('-')
    start.append(int(s_year))
    end.append(int(e_year))

odi_bat_df['Start_year_ODI_bat'] = start
odi_bat_df['End_year_ODI_bat'] = end
odi_bat_df['Span_ODI_bat'] = odi_bat_df['End_year_ODI_bat'] - odi_bat_df['Start_year_ODI_bat']

odi_bat_df
```

#### Handling `HS` columns
The `*` in the score shows that the player was **Not out** in that inning. We will make a separate feature for that named `HS_ODI_isNO` to see if the player was not out or not. Then we can remove the `*` from the HS

```python
is_NO = []
hs_new = []
for hs in odi_bat_df['HS_ODI']:
    if(hs[-1] == '*'):
        is_NO.append(1)
        hs_new.append(int(hs[:-1]))
    else:
        is_NO.append(0)
        hs_new.append(int(hs)) 

odi_bat_df['HS_ODI'] = hs_new
odi_bat_df['HS_ODI_isNO'] = is_NO

odi_bat_df
```

#### Now we have successfully handled all the issues in the data set of ODI batting
